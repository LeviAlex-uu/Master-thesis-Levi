```python
import os

path = os.getcwd()

file = open('Labels.txt', 'r')
labels = file.readlines()
```

```python
labels = [x.replace('\n', '').lower() for x in labels]
print(labels)
```

```python
print(len(labels))
print(len(set(labels)))
```

```python
import numpy as np
import matplotlib.pyplot as plt

values, counts = np.unique(labels, return_counts=True)
count_dict = dict(zip(values, counts))

N = 20
res = dict(sorted(count_dict.items(), key = lambda x: x[1], reverse=True)[:N])

fig, ax = plt.subplots()
ax.bar(res.keys(), res.values())
ax.set_ylim(0,20)
plt.xticks(rotation=45, ha='right')
```
