```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.metrics as skl
```

```python
def castMatrix(data, dtype=int):
    return list(map(lambda sub: list(map(dtype, sub)), data))

def translation(x):
    dict = {'noMatch' : 0, 0 : 'noMatch', 'relatedMatch' : 1, 1 : 'relatedMatch', 'exactMatch' : 2, 2 : 'relatedMatch'}
    return dict[x]
```

```python
accs = list(map(int, np.loadtxt('results_accs.txt')))
halfaccs = list(map(int, np.loadtxt('results_halfaccs.txt')))
true_predicates = list(map(translation, np.loadtxt('results_true_predicates.txt')))
predicates = castMatrix(np.loadtxt('results_predicates.txt'), translation)
mrrs = list(map(int, np.loadtxt('results_halfaccs.txt')))
```

```python
models = ['Levenshtein','LCS','Word2Vec']
cutoffs = np.arange(1,10,1)/10
N = 574
cutoffs
```

```python
true_acc = []
g_acc = []
mrr_score = []

for i,_ in enumerate(models):
    true_acc.append(accs[i::3])
    g_acc.append(halfaccs[i::3])
    mrr_score.append(mrrs[i::3])
```

```python
fig,axs = plt.subplots(3)
for index,i in enumerate([true_acc,g_acc,mrr_score]):
    axs[index].plot(cutoffs,[round(j/N,2)*100 for j in i[0]], label='Levenshtein', linestyle = 'solid')
    axs[index].plot(cutoffs,[round(j/N,2)*100 for j in i[1]], label='Lcs', linestyle = 'dotted')
    axs[index].plot(cutoffs,[round(j/N,2)*100 for j in i[2]], label='Word2Vec', linestyle = 'dashed')
    axs[index].set(xlabel='cutoff', ylabel='accuracy (%)')
    axs[index].set_yticks(np.arange(0, 31, 5))
    axs[index].grid()
    axs[index].legend()

plt.show()
```
